```python
import pandas as pd
import matplotlib.pyplot as plt
from mplsoccer.pitch import Pitch
import seaborn as sns
```

```python
df = pd.read_csv('vsLethbridge_1stHalf.csv').dropna().reset_index(drop=True)
df_2 = pd.read_csv('vsLethbridge_2ndHalf.csv').dropna().reset_index(drop=True)
df_full = pd.concat([df, df_2], ignore_index=True)
```

```python
df['X'] = df['X']*1.2
df['Y'] = df['Y']*.8
df['X2'] = df['X2']*1.2
df['Y2'] = df['Y2']*.8

df_2['X'] = df_2['X']*1.2
df_2['Y'] = df_2['Y']*.8
df_2['X2'] = df_2['X2']*1.2
df_2['Y2'] = df_2['Y2']*.8

df_full['X'] = df_full['X']*1.2
df_full['Y'] = df_full['Y']*.8
df_full['X2'] = df_full['X2']*1.2
df_full['Y2'] = df_full['Y2']*.8
```

```python
pitch = Pitch(pitch_type='statsbomb', pitch_color='#22312b', line_color='#c7d5cc')
fig, ax = pitch.draw(figsize=(13.5, 8), constrained_layout=True, tight_layout=False)
pitch.draw(ax=ax)
plt.gca().invert_yaxis()
plt.gca().invert_xaxis()

for x in range(len(df['X'])):
    if df['Event'] [x] == 'Pass Yes':
        start_x, start_y = df['X'][x], df['Y'][x]
        end_x, end_y = df['X2'][x], df['Y2'][x]
        
        successful = plt.arrow(start_x, start_y, end_x - start_x, end_y - start_y,
                  head_width=1, head_length=1, fc='green', ec='green')

    if df['Event'] [x] == 'Pass No':
        
        start_x, start_y = df['X'][x], df['Y'][x]
        end_x, end_y = df['X2'][x], df['Y2'][x]
        
        unsuccessful = plt.arrow(start_x, start_y, end_x - start_x, end_y - start_y,
                  head_width=1, head_length=1, fc='red', ec='red')

successful.set_label('successful pass')
unsuccessful.set_label('unsuccessful pass')
ax.legend(facecolor='#22312b', handlelength=5, edgecolor='None', fontsize=10, loc='upper left', bbox_to_anchor=(0.02, 1.10))

plt.title('Pass Map - First Half VS Lethbridge \n -------------------------------------> \n Attacking')
```
